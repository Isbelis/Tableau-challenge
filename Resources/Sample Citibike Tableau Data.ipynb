```python
import os
import pandas as pd
```

```python
# init main df
main_df = pd.DataFrame()

# loop through all sub folders
for subdir, dirs, files in os.walk("C:\\Users\\abooth\\Documents\\SMU_2023\\Tableau Data Files\\2023-citibike-tripdata"):
    for file in files:
        filepath = os.path.join(subdir, file)
        
        # check if CSV data file
        if (".csv" in filepath):
            # make sample df
            df = pd.read_csv(filepath, low_memory=False)
            df = df.sample(frac=0.01, random_state=42)
            main_df = pd.concat([main_df, df], axis=0)
            print(filepath)
    
main_df = main_df.reset_index(drop=True)
```
